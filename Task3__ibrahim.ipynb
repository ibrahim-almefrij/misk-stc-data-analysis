```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_excel('stc TV Data Set_T3.xlsx', sheet_name='Sheet1')
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df = df.dropna(subset=['rating'])
```

```python
pivot = df.pivot_table(index='user_id_maped', columns='program_name', values='rating').fillna(0)
sim = pd.DataFrame(cosine_similarity(pivot.T), index=pivot.columns, columns=pivot.columns)
```

```python
print("Top 5 Recommendations for Moana:")
print(sim['Moana'].sort_values(ascending=False).iloc[1:6])
```

```python
print("\nSample Results:")
print(sim.iloc[:5, :5])
```
